```python
import mne
import os
import numpy as np
import pandas as pd
import scipy.io
from scipy.signal import butter, filtfilt, welch, windows
from scipy import signal
import matplotlib.pyplot as plt
```

```python
band = 'delta'
#delta: 0.1-4
#theta: 4-7
#alpha 8-12
cluster = 5
```

### CSZ Labelling

```python
folder = 'E:/sayan_24/GNN_aniket/CSZ_select_' + band + '_cluster_' + str(cluster) + '/'
save_dir = 'E:/sayan_24/GNN_aniket/CSZ_label_data_' + band + '_cluster_' + str(cluster) + '/'
os.makedirs(save_dir, exist_ok = True)

for case in range(1, 24):
    df_csz = scipy.io.loadmat(folder + f'train_csz{case}.mat')
    df_test = scipy.io.loadmat(folder + f'test{case}.mat')
    df_he = scipy.io.loadmat(folder + 'train_he_full.mat')

    df_csz = df_csz['train']
    df_test = df_test['test']
    df_he = df_he['train']

    csz = np.array([0, 1])
    test = np.array([0, 1])
    he = np.array([1, 0])

    label_csz = np.tile(csz, (len(df_csz), 1))
    label_test = np.tile(test, (len(df_test), 1))
    label_he = np.tile(he, (len(df_he), 1))

    
    save_path = os.path.join(save_dir, f'csz_label{case}.mat')
    scipy.io.savemat(save_path, {'csz_label': label_csz}) 

    save_path = os.path.join(save_dir, f'test_label{case}.mat')
    scipy.io.savemat(save_path, {'test_label': label_test})

    save_path = os.path.join(save_dir, f'he_label{case}.mat')
    scipy.io.savemat(save_path, {'he_label': label_he}) 
```

### HE Labelling

```python
folder = 'E:/sayan_24/GNN_aniket/HE_select_' + band + '_cluster_' + str(cluster) + '/'
save_dir = 'E:/sayan_24/GNN_aniket/HE_label_data_' + band + '_cluster_' + str(cluster) + '/'
os.makedirs(save_dir, exist_ok = True)

for case in range(1, 41):
    df_he = scipy.io.loadmat(folder + f'train_he{case}.mat')
    df_test = scipy.io.loadmat(folder + f'test{case}.mat')
    df_csz = scipy.io.loadmat(folder + 'train_csz_full.mat')

    df_csz = df_csz['train']
    df_test = df_test['test']
    df_he = df_he['train']

    csz = np.array([0, 1])
    test = np.array([1, 0])
    he = np.array([1, 0])

    label_csz = np.tile(csz, (len(df_csz), 1))
    label_test = np.tile(test, (len(df_test), 1))
    label_he = np.tile(he, (len(df_he), 1))


    save_path = os.path.join(save_dir, f'csz_label{case}.mat')
    scipy.io.savemat(save_path, {'csz_label': label_csz}) 

    save_path = os.path.join(save_dir, f'test_label{case}.mat')
    scipy.io.savemat(save_path, {'test_label': label_test})

    save_path = os.path.join(save_dir, f'he_label{case}.mat')
    scipy.io.savemat(save_path, {'he_label': label_he}) 
```
