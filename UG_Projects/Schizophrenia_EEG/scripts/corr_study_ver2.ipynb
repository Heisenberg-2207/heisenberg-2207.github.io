```python
import mne
import os
import numpy as np
import pandas as pd
import scipy.io
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
```

```python
signal_strength = 0.85
band = 'delta'
#delta: 0.1-4
#theta: 4-7
#alpha 8-12
cluster = 5
```

### CSZ data

```python
csz_folder = 'E:/sayan_24/GNN_aniket/CSZ_select_' + band + '_cluster_' + str(cluster) + '/'
save_dir = 'E:/sayan_24/GNN_aniket/CSZ_corr_data_' + band + '_cluster_' + str(cluster) + '/'
os.makedirs(save_dir, exist_ok = True)


for case in range(1, 24):
    df_csz = scipy.io.loadmat(csz_folder + f'train_csz{case}.mat')
    df_test = scipy.io.loadmat(csz_folder + f'test{case}.mat')
    df_he = scipy.io.loadmat(csz_folder+ 'train_he_full.mat')

    df_csz_np = np.array(df_csz['train'])
    df_test_np = np.array(df_test['test'])
    df_he_np = np.array(df_he['train'])

    csz_corr_mat =[]
    test_corr_mat = []
    he_corr_mat = []

    for mat in df_csz_np:
        corr_matrix = np.corrcoef(mat, rowvar=False)
        np.fill_diagonal(corr_matrix, 0)
        corr_matrix = corr_matrix*signal_strength
        csz_corr_mat.append(corr_matrix)
    

    for mat in df_test_np:
        corr_matrix = np.corrcoef(mat, rowvar=False)
        np.fill_diagonal(corr_matrix, 0)
        corr_matrix = corr_matrix*signal_strength
        test_corr_mat.append(corr_matrix)

    for mat in df_he_np:
        corr_matrix = np.corrcoef(mat, rowvar=False)
        np.fill_diagonal(corr_matrix, 0)
        he_corr_mat.append(corr_matrix)


    # Save csz_corr_mat as .mat 
    save_path = os.path.join(save_dir, f'csz_corr_matrix{case}.mat')
    scipy.io.savemat(save_path, {'csz_corr_mat': csz_corr_mat})

    # Save corr_matrix as .mat file
    save_path = os.path.join(save_dir, f'test_corr_matrix{case}.mat')
    scipy.io.savemat(save_path, {'corr_matrix': test_corr_mat})

    # Save he_corr_mat as .mat file
    save_path = os.path.join(save_dir, f'he_corr_matrix{case}.mat')
    scipy.io.savemat(save_path, {'he_corr_mat': he_corr_mat})
```
