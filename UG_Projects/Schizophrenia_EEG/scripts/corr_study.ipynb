```python
import mne
import os
import numpy as np
import pandas as pd
import scipy.io
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
```

```python
csz_folder = 'D:/heisen/IITM studies/GNN/CSZ_select/'

for case in range(1, 2):
    df_csz = scipy.io.loadmat(csz_folder + f'train_csz{case}.mat')
    df_test = scipy.io.loadmat(csz_folder + f'test{case}.mat')
    df_he = scipy.io.loadmat(csz_folder+ 'train_he_full.mat')

    df_csz_np = np.array(df_csz['train'])
    df_test_np = np.array(df_test['test'])
    df_he_np = np.array(df_he['train'])

    csz_corr_mat =[]
    test_corr_mat = []
    he_corr_mat = []

    for mat in df_csz_np:
        corr_matrix = np.corrcoef(mat, rowvar=False)
        csz_corr_mat.append(corr_matrix)

    for mat in df_test_np:
        corr_matrix = np.corrcoef(mat, rowvar=False)
        test_corr_mat.append(corr_matrix)

    for mat in df_he_np:
        corr_matrix = np.corrcoef(mat, rowvar=False)
        he_corr_mat.append(corr_matrix)

    
    # Save csz_corr_mat as .mat file
    scipy.io.savemat(f'csz_corr_matrix{case}.mat', {'csz_corr_mat': csz_corr_mat})

    # Save corr_matrix as .mat file
    scipy.io.savemat(f'test_corr_matrix{case}.mat', {'corr_matrix': test_corr_mat})

    # Save he_corr_mat as .mat file
    scipy.io.savemat(f'he_corr_matrix{case}.mat', {'he_corr_mat': he_corr_mat})
```
