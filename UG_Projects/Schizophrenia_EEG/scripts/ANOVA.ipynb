```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Load the DataFrame
df = pd.read_csv('table_EC.csv', header=None)

# Remove the square brackets from the values in the first column
df[1] = df[1].str.replace('[', '').str.replace(']', '')

# Display the modified DataFrame
df.head()
```

```python
# Separate DataFrame based on the condition
df_csz = df[df.iloc[:, 0].str.contains('CSZ', na=False)].iloc[:, [0, 1]]
df_he = df[df.iloc[:, 0].str.contains('HE', na=False)].iloc[:, [0, 1]]

# Display the resulting DataFrames
print("DataFrame with 'CSZ':")
print(df_csz)

print("\nDataFrame with 'HE':")
print(df_he)
```

```python
# Split the comma-separated values in the first column into separate columns
df_csz_split = df_csz[1].str.split(',', expand=True)
df_he_split = df_he[1].str.split(',', expand=True)

# Add the second column back to the split DataFrames
df_csz_split.insert(0, 'Type of Cluster and Band', df_csz[0])
df_he_split.insert(0, 'Type of Cluster and Band', df_he[0])

# Rename the columns
df_csz_split.columns = ['Type of Cluster and Band'] + [f'Patient {i}' for i in range(1, df_csz_split.shape[1])]
df_he_split.columns = ['Type of Cluster and Band'] + [f'Patient {i}' for i in range(1, df_he_split.shape[1])]
```

```python
# Add a 'Group' column to each DataFrame
df_csz_split['Group'] = 'CSZ'
df_he_split['Group'] = 'HE'

# Combine the two DataFrames
combined_df = pd.concat([df_csz_split, df_he_split], ignore_index=True)

# Melt the combined DataFrame to have a long format
melted_df = combined_df.melt(id_vars=['Type of Cluster and Band', 'Group'], var_name='Patient', value_name='Value')

# Convert the 'Value' column to numeric, coercing errors to NaN
melted_df['Value'] = pd.to_numeric(melted_df['Value'], errors='coerce')

# Drop rows with NaN values in the 'Value' column
melted_df = melted_df.dropna(subset=['Value'])

# Perform ANOVA
model = ols('Value ~ C(Group)', data=melted_df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display the ANOVA table
print(anova_table)
```
